```python
# Load UCI census train and test data into dataframes.
import pandas as pd
features = ["Team", "Sprint #", "Start Date", "End Date", "Working Days", "Leaves", "Story Points", "Time Taken (Mins)", "All Committed Storypoints", "All Committed User Stories", "All Added Later", "All Added Later Stories", "All Actual Storypoints (Velocity)", "All Completed Stories", "All Total Storypoints", "All Removed Story Points", "All Average Velocity", "Spillover - All", "All Accuracy", "All Vertical Line 1", "Cyb Committed Storypoints", "Cyb Committed User Stories", "Cyb Added Later", "Cyb Added Later Stories", "Cyb Actual Storypoints (Velocity)", "Cyb Completed Stories", "Cyb Total Storypoints", "Cyb Removed Story Points", "Cyb Average Velocity", "Spillover - Cyb", "Cyb Accuracy", "Cyb Vertical Line 2", "US Committed Storypoints", "US Added Later", "US Actual Storypoints (Velocity)", "US Total Storypoints", "US Removed Story Points", "US Average Velocity", "US Spillover - US", "US Accuracy", "US Vertical Line 3", "Cyb Team Size", "Cyb Available Hours", "Cyb Burnt Hours", "Cyb Productivity"]
train_data = pd.read_csv(
    "C:\\Users\\dhavalma\\facets\\Custom_Project_Data-Filebound.csv",
    names=features,
    sep=r'\s*,\s*',
    engine='python',
    na_values="?")
test_data = pd.read_csv(
    "C:\\Users\\dhavalma\\facets\\Custom_Project_Data-Filebound.csv",
    names=features,
    sep=r'\s*,\s*',
    skiprows=[0],
    engine='python',
    na_values="?")
```

```python
# Calculate the feature statistics proto from the datasets and stringify it for use in facets overview
from generic_feature_statistics_generator import GenericFeatureStatisticsGenerator
import base64

gfsg = GenericFeatureStatisticsGenerator()
proto = gfsg.ProtoFromDataFrames([{'name': 'train', 'table': train_data},
                                  {'name': 'test', 'table': test_data}])
protostr = base64.b64encode(proto.SerializeToString()).decode("utf-8")
```

```python
# Display the facets overview visualization for this data
from IPython.core.display import display, HTML

HTML_TEMPLATE = """<link rel="import" href="/nbextensions/facets-dist/facets-jupyter.html" >
        <facets-overview id="elem"></facets-overview>
        <script>
          document.querySelector("#elem").protoInput = "{protostr}";
        </script>"""
html = HTML_TEMPLATE.format(protostr=protostr)
display(HTML(html))
```
